```python
import pandas as pd
from scipy.stats import zscore 
```

```python
data = pd.read_csv("prices_round_2_day_0.csv", sep=';')
```

```python
data
```

```python
import matplotlib.pyplot as plt
```

```python
df = data
```

```python
df['SUNLIGHT_zscore'] = zscore(df['SUNLIGHT'])
df['HUMIDITY_zscore'] = zscore(df['HUMIDITY'])
df['IMPORT_TARIFF_zscore'] = zscore(df['IMPORT_TARIFF'])
df['EXPORT_TARIFF_zscore'] = zscore(df['EXPORT_TARIFF'])
df['TRANSPORT_FEES_zscore'] = zscore(df['TRANSPORT_FEES'])
df['ORCHIDS_zscore'] = zscore(df['ORCHIDS'])
```

```python
df
```

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed

# Plot parameter 1
plt.plot(df['timestamp'], df['SUNLIGHT_zscore'], label='Sunlight')

# Plot parameter 2
plt.plot(df['timestamp'], df['HUMIDITY_zscore'], label='Humidity')


plt.plot(df['timestamp'], df['IMPORT_TARIFF_zscore'], label='Import')
plt.plot(df['timestamp'], df['EXPORT_TARIFF_zscore'], label='Export')
plt.plot(df['timestamp'], df['TRANSPORT_FEES_zscore'], label='Transport')
plt.plot(df['timestamp'], df['ORCHIDS_zscore'], label='Orchids')

# Add labels and title
plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.title('Sunlight and Humidity Over Time')

# Add legend
plt.legend()

# Show plot
plt.show()
```
